## Custom Graph

```python
# values on x & y axis
names = ['Asnan','Hamza','Saif','Talha']
gpa = [3.3,2.94,2.5,2.6]
```

```python
# title
plt.title("CGPA 4th Semester")

# label the axis
plt.xlabel("Names")
plt.ylabel("CGPA")

# plotting the graph
plt.plot(names, gpa)
```

```python
# other attributes works the same way
plt.scatter(names, gpa)
```

```python
plt.bar(names, gpa)
```
