## Multiplot on same canvas

```python
# plt.subplot(rows, columns, plot_number)
# meaning, 1 row will have 2 column(plots)

plt.subplot(1,3,1) # first plot
plt.plot(x, y, 'r')

plt.subplot(1,3,2) # second plot
plt.plot(y, x, 'g')

plt.subplot(1,3,3) # third plot
plt.bar(x,y)
```

```python
# reversing the plot into rows
# 3 rows & 1 column each
plt.subplot(3,1,1)
plt.plot(x,y,'r')

plt.subplot(3,1,2)
plt.plot(y,x,'g')

plt.subplot(3,1,3)
plt.plot(x,y)
```

```python
# dotted graph
plt.subplot(1,2,1) # first plot
plt.plot(x, y, 'r--') # dotted line

plt.subplot(1,2,2) # second plot
plt.plot(y, x, 'go-') # circular line ( '*-' can also be used)
```

```python
plt.plot(x, y, 'ro')
```

```python
plt.plot(y, x, 'g*')
```

## Matplotlib Object Oriented Method

As the name is self-explanatory, we will create object and then use that object to call various methods and attributes as in general concept of OOP

Basic Steps:

- Create figure()
- add axes
- plot graphs using axes

```python
# Creating Object (empy canvas)
fig = plt.figure()

# adding axes (we will use this axes to add graph into it)
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (position, scale)

# plotting graph
axes.plot(x,y,'g--')
```

```python
# plotting multiple graph in same canvas
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height
axes.plot(x,y,'g')
axes.plot(y,x,'r')
axes.plot(y,y,'b')
```

#### Changes in commands while using 'axes'

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height
axes.plot(x,y,'g')
axes.plot(y,x,'r')

# changes 'set' is being used here
axes.set_xlabel('x values')
axes.set_ylabel('y values')
axes.set_title("This is title")
```

### Axes within Axes

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # left, bottom, width, height

# larger figure
axes1.plot(x,y,'g')
axes1.set_title('1st Axes')

# smaller figure
axes2.plot(y,x,'r--')
axes2.set_title("2nd Axes")
```

## Line Plot

```python
import matplotlib.pyplot as plt

# x axis
x = [1,2,3]
# y axis
y = [2,1,4]

# plotting the graph
plt.plot(x,y,'g')

# title
plt.title("Line Graph")
```

### Multi-line plot

```python
# first line
x = [1,2,3]
y = [2,1,4]

# second line
x1 = [2,3,5]
y1 = [6,1,3]

# plotting the graph
plt.plot(x,y,'g',label='Green Line')
plt.plot(x1,y1,'r',label='Red Line')

# title
plt.title("Multi Lines")
plt.xlabel("x axis")
plt.ylabel("y axis")

# legend
plt.legend()
```

## Pie Chart

Note: You need to pass labels here. Which is a must!

```python
# functional method
langs = ['C++','Python','Dart','Java']
std = [20,40,34,22]

plt.pie(std, labels=langs, autopct='%1.2f%%') # label are used as languages
# %1.2f%% is used to make sure values don't exceeds 2 decimal places
```

```python
# object oriented method
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])

langs = ['C++','Python','Dart','Java']
std = [20,40,34,22]

ax.pie(std, labels=langs, autopct='%1.2f%%')
```

# Real World Example

```python
# Taking datasets from iris
from sklearn.datasets import load_iris

iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2, s = 100 * features[3], c=iris.target, cmap='viridis')

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```
