# Matplotlib Part 2

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
%matplotlib inline
````

````python
x = np.linspace(0,5, 11)
````

````python
fig, axes = plt.subplots(1, 2, figsize = (10, 4))
# axes[axes_number].plot(x1, y1, x2, y2....)
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal Scale")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logrithmic Scale (y)")
````

### Placement of ticks & Custom ticks
We can use explicit functions to customize the values of the tickes i.e. ```set_xticks``` and ```set_yticks``` which takes list of values for where the ticks are to be placed on the axis. Similarly, we can use ```set_xticklabels``` and ```set_yticklabels``` to set the label for them.

````python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(x, x**2, x, x**3, lw=2)

# passing arring into method
ax.set_xticks = ([1,2,3,4,5])
# passing list variable
ax.set_xticklabels([r'$\alpha$', r'$\delta$', r'$\beta$', r'$\gamma$', r'$\phi$', r'$\sigma$', r'$\epsilon$'], fontsize=18)

# defining a variable and then passing it later on
yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.1f$" % y for y in yticks]) # label at one decimal places
````

## Setting the padding of x & y label

````python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(x, x**2, x, np.exp(x), lw=2)

# passing arring into method
ax.set_xticks = ([1,2,3,4,5])
# passing list variable
ax.set_xticklabels([r'$\alpha$', r'$\delta$', r'$\beta$', r'$\gamma$', r'$\phi$', r'$\sigma$', r'$\epsilon$'], fontsize=18)

# defining a variable and then passing it later on
yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.1f$" % y for y in yticks]) # label at one decimal places

# setting the label padding
ax.xaxis.labelpad = 20
ax.yaxis.labelpad = 30

ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
````

## Adjustment of Axis

````python
fig, ax = plt.subplots(1,1)

ax.plot(x, x**2, x**2, x)
ax.set_title("Adjustment of Axis")

ax.set_yticks([10,20,50,200])

# figure axis position from LRBT
# comment belo line to see the difference
fig.subplots_adjust(left=0.0, right=0.9, bottom=0.1, top=0.9)
````

## Axis Grid
Default & Custom Grid

````python
fig, ax = plt.subplots(1, 2, figsize=(10,2))

ax[0].plot(x, x**2, 'r')
ax[1].plot(x**2, x, 'g')

# default grid
ax[0].grid(True)
ax[0].set_title("Default Grid")

# custom grid
ax[1].set_title("Custom Grid")
# axes[axes_index] = (color= , alpha= , linestyle= , linewidth= )
ax[1].grid(color='r', alpha=0.7, linestyle='dashed', linewidth=0.9)
````

## Axis Spines
The border lines are known as spine

````python
fig, ax = plt.subplots(1, 2, figsize=(10,2))

ax[0].plot(x, x**2, 'r')
ax[1].plot(x**2, x, 'g')

# changing the color
# axes[axes_index].spines['position_string'].set_color('color_string')
ax[0].spines['bottom'].set_color('blue')
ax[1].spines['right'].set_color('red')

# making the axis transparent
ax[1].spines['top'].set_color('none')

# changing the width
# axes[axes_index].spines['position_string'].set_linewidth(double)
ax[0].spines['top'].set_linewidth(2.8)
````

## Twin Axes
Helps us in getting double x or y axis on both sides of the graph. Matplotlib supports this with functions line ```twinx``` and ```twiny```.

````python
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color='red')
# setting up the label for square
ax1.set_ylabel(r'area $(m^2)$', fontsize=18, color="red")
# changing color of y_ticks
for label in ax1.get_yticklabels():
    label.set_color("red")

# twin axis
ax2 = ax1.twinx()

ax2.plot(x, x**3, lw=2, color='green')
# setting up the label for cube
ax2.set_ylabel(r'area $(m^3)$', fontsize=28, color="green")
# color of y_ticks accordingly
for label in ax2.get_yticklabels():
    label.set_color("green")
````

## Axes with x & y are zero

````python
fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
# set position of x spine to x=0
ax.spines['bottom'].set_position(('data', 0))

ax.yaxis.set_ticks_position('left')
# set position of y spine to y=0
ax.spines['left'].set_position(('data', 0))

xx = np.linspace(-0.75, 1, 100)
ax.plot(xx, xx**3, 'r--')
````

## Other 2D Plot type
Check out other ```plot``` using Matplotlib gallery at http://matplotlib.org/gallery.html

````python
x = np.array([1, 2, 3, 4, 5])
````

````python
fig, ax = plt.subplots(1, 4, figsize=(18,3))

ax[0].scatter(x, np.random.randn(len(x)), facecolor='green')
ax[0].set_title("Scatter")

ax[1].step(x, x**2, 'r--')
ax[1].set_title("Step")

ax[2].bar(x, x**2, width=0.5)
ax[2].set_title("Bar")

ax[3].fill_between(x, x**2, x**3, facecolor='green', alpha=0.4)
ax[3].set_title("Fill Between")
````

## Text Annotation
Annotating text to matplotlib figures can be done using ```text``` function

````python
fib, ax = plt.subplots()

ax.plot(x, x**2, x**3, x**2)
ax.text(20, 5, r'$y=x^2$', fontsize=18, color='green')
ax.text(10, 21, r'$y=x^3$', fontsize=18, color='red')
````
